## 步驟3：資料清理

```python
import string
import pandas as pd
import nltk
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords
import re
import matplotlib.pyplot as plt
import seaborn as sns

# 初始化 TweetTokenizer
tokenizer = TweetTokenizer()

# 修改停用詞，保留否定詞
stop_words = set(stopwords.words('english'))
stop_words.difference_update({'not', 'no', 'never'})  # 保留否定詞

# 修改標點符號處理，保留 ! 和 ?
punctuation = string.punctuation.replace('!', '').replace('?', '')

# 定義資料清理的函式（修正標點符號處理）
def clean_text(text):
    # 1. 去除 URL
    text = re.sub(r'http\S+|www.\S+', '', text)
    # 2. 轉小寫
    text = text.lower()
    # 3. 去除部分標點符號（保留 ! 和 ?）
    text = text.translate(str.maketrans('', '', punctuation))
    # 4. 分詞
    words = tokenizer.tokenize(text)
    # 5. 去除停用詞（保留否定詞）
    words = [word for word in words if word not in stop_words]
    # 6. 移除非英文字（如數字、符號等），保留 ! 和 ?
    words = [word for word in words if word.isalpha() or word in ['!', '?']]
    return words

# 對訓練與測試資料的留言欄位進行清理與分詞
train_df['clean_tokens'] = train_df['OriginalTweet'].apply(clean_text)
test_df['clean_tokens'] = test_df['OriginalTweet'].apply(clean_text)

# 計算每個樣本的 token 長度
train_df['token_length'] = train_df['clean_tokens'].apply(len)
test_df['token_length'] = test_df['clean_tokens'].apply(len)

# 顯示清理後的結果（包含長度）
print("清理後的訓練資料前 5 筆：")
print(train_df[['OriginalTweet', 'clean_tokens', 'token_length']].head())
print("\n清理後的測試資料前 5 筆：")
print(test_df[['OriginalTweet', 'clean_tokens', 'token_length']].head())

# 統計長度分佈
print("\n訓練集 token 長度統計：")
print(train_df['token_length'].describe())
print("\n測試集 token 長度統計：")
print(test_df['token_length'].describe())

# 視覺化長度分佈
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(train_df['token_length'], bins=30, kde=True)
plt.title('訓練集 Token 長度分佈')
plt.xlabel('Token 數量')
plt.ylabel('頻率')

plt.subplot(1, 2, 2)
sns.histplot(test_df['token_length'], bins=30, kde=True)
plt.title('測試集 Token 長度分佈')
plt.xlabel('Token 數量')
plt.ylabel('頻率')

plt.tight_layout()
plt.savefig('token_length_distribution.png')
plt.show()
```

## 步驟4：GloVe 詞向量化和嵌入矩陣生成

```python
# 步驟4：GloVe Twitter 分詞與向量化
# 定義函式：從本地檔案載入GloVe Twitter嵌入向量
def load_glove_vectors(glove_file):
    word_vectors = {}
    with open(glove_file, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.strip().split()
            word = values[0]
            vector = np.array(values[1:], dtype='float32')
            word_vectors[word] = vector
    return word_vectors

# 載入GloVe Twitter嵌入向量（假設使用100維）
GLOVE_FILE = r'C:\Users\skych\glove.twitter.27B\glove.twitter.27B.100d.txt'
glove_vectors = load_glove_vectors(GLOVE_FILE)

# 創建詞彙表
all_tokens = [token for tokens in train_df['clean_tokens'] for token in tokens]
vocab = {word: idx + 1 for idx, word in enumerate(set(all_tokens))}  # +1 為padding保留索引0
vocab['<PAD>'] = 0  # 添加padding token
VOCAB_SIZE = len(vocab)

MAX_LENGTH = 50

# 定義函式：將清理後的tokens轉換為索引
def tokens_to_indices(tokens, vocab, max_length=MAX_LENGTH):
    indices = [vocab.get(token, vocab['<PAD>']) for token in tokens]
    if len(indices) < max_length:
        indices += [vocab['<PAD>']] * (max_length - len(indices))  # 填充
    return indices[:max_length]  # 截斷至最大長度

# 將訓練與測試資料的tokens轉為索引
train_df['input_indices'] = train_df['clean_tokens'].apply(lambda x: tokens_to_indices(x, vocab))
test_df['input_indices'] = test_df['clean_tokens'].apply(lambda x: tokens_to_indices(x, vocab))

# 定義函式：生成GloVe Twitter嵌入矩陣
def get_embedding_matrix(vocab, glove_vectors, embedding_dim):
    embedding_matrix = np.zeros((len(vocab), embedding_dim))  # 詞彙表大小包含padding
    for word, idx in vocab.items():
        if word in glove_vectors and word != '<PAD>':
            embedding_matrix[idx] = glove_vectors[word]
        else:
            # 為OOV詞或padding token隨機初始化嵌入向量
            embedding_matrix[idx] = uniform_(torch.empty(embedding_dim)).numpy()
    return torch.tensor(embedding_matrix, dtype=torch.float32)

# 生成GloVe Twitter嵌入矩陣
EMBEDDING_DIM = 100  # GloVe Twitter 100維
embedding_matrix = get_embedding_matrix(vocab, glove_vectors, EMBEDDING_DIM)

# 顯示詞向量化後的結果
print("詞向量化後的訓練資料前5筆：")
print(train_df[['clean_tokens', 'input_indices']].head())
```

## 步驟5.6：網格搜索與最佳組合選擇

```python
## 步驟5.6：網格搜索與最佳組合選擇
import nltk
from nltk.tokenize import TweetTokenizer
from itertools import product
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.optim.lr_scheduler import ReduceLROnPlateau
from tqdm import tqdm
import pandas as pd
import numpy as np
import re
import string
from nltk.corpus import stopwords
from sklearn.utils.class_weight import compute_class_weight
import os

# 下載NLTK數據
nltk.download('stopwords', quiet=True)
nltk.download('punkt', quiet=True)

# 初始化TweetTokenizer
tokenizer = TweetTokenizer()

# 定義縮寫字典
ABBREVIATIONS = {
    'lol': 'laugh out loud',
    'brb': 'be right back',
    'omg': 'oh my god',
    'wtf': 'what the fuck',
    'lmao': 'laughing my ass off',
    'rofl': 'rolling on the floor laughing',
    'idk': 'i do not know',
    'smh': 'shaking my head',
    'btw': 'by the way',
    'fyi': 'for your information',
    'asap': 'as soon as possible'
}

# 定義載入GloVe詞向量的函式
def load_glove_vectors(glove_file):
    if not os.path.exists(glove_file):
        raise FileNotFoundError(f"GloVe file not found at {glove_file}")
    glove_vectors = {}
    with open(glove_file, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.strip().split()
            word = values[0]
            vector = np.array(values[1:], dtype='float32')
            glove_vectors[word] = vector
    return glove_vectors

# 修改資料清理函式，支援動態參數
def clean_text(text, expand_abbreviations=False, keep_usernames=False, keep_negation=True):
    text = re.sub(r'http\S+|www.\S+', '', text)
    text = text.lower()
    punctuation = string.punctuation.replace('!', '').replace('?', '')
    text = text.translate(str.maketrans('', '', punctuation))
    words = tokenizer.tokenize(text)
    if expand_abbreviations:
        words = [ABBREVIATIONS.get(word, word) for word in words]
        words = [subword for word in words for subword in tokenizer.tokenize(word)]
    if not keep_usernames:
        words = [word for word in words if not word.startswith('@')]
    else:
        words = [word[1:] if word.startswith('@') else word for word in words]
    stop_words = set(stopwords.words('english'))
    if keep_negation:
        stop_words.difference_update({'not', 'no', 'never'})
    words = [word for word in words if word not in stop_words]
    words = [word for word in words if word.isalpha()]
    return words

# 定義 tokens_to_indices 函式
def tokens_to_indices(tokens, vocab, max_length):
    indices = [vocab.get(token, vocab['<PAD>']) for token in tokens]
    if len(indices) < max_length:
        indices += [vocab['<PAD>']] * (max_length - len(indices))
    return indices[:max_length]

# 定義 get_embedding_matrix 函式
def get_embedding_matrix(vocab, glove_vectors, embedding_dim):
    embedding_matrix = torch.zeros((len(vocab), embedding_dim))
    for word, idx in vocab.items():
        if word in glove_vectors:
            embedding_matrix[idx] = torch.tensor(glove_vectors[word], dtype=torch.float)
        else:
            embedding_matrix[idx] = torch.randn(embedding_dim)
    return embedding_matrix

# 高性能 BiLSTMClassifier
class BiLSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim, num_layers=4, dropout=0.2, use_hidden_states=False, use_attention=False, embedding_matrix=None):
        super(BiLSTMClassifier, self).__init__()
        self.hidden_dim = hidden_dim
        self.use_attention = use_attention
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        if embedding_matrix is not None:
            self.embedding.weight = nn.Parameter(embedding_matrix)
            self.embedding.weight.requires_grad = True
        self.dropout = nn.Dropout(dropout)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers=num_layers, 
                           bidirectional=True, batch_first=True, dropout=dropout if num_layers > 1 else 0)
        if use_attention:
            self.attention = nn.Linear(hidden_dim * 2, 1)
        self.batch_norm = nn.BatchNorm1d(hidden_dim * 2)
        self.fc = nn.Linear(hidden_dim * 2, output_dim)
        self.use_hidden_states = use_hidden_states

    def forward(self, text):
        embedded = self.dropout(self.embedding(text))  # [batch_size, max_length, embedding_dim]
        lstm_out, (hidden, cell) = self.lstm(embedded)  # lstm_out: [batch_size, max_length, hidden_dim*2]
        if self.use_attention:
            attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
            context = torch.sum(attn_weights * lstm_out, dim=1)
            hidden = context
        elif self.use_hidden_states:
            hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)
        else:
            hidden = torch.mean(lstm_out, dim=1)
        hidden = self.batch_norm(hidden)
        hidden = self.dropout(hidden)
        return self.fc(hidden)

# 修改 CommentDataset，支援 One-Hot 標籤
class CommentDataset(Dataset):
    def __init__(self, input_indices, labels, one_hot=False):
        self.input_indices = input_indices
        self.one_hot = one_hot
        if one_hot:
            encoder = OneHotEncoder(sparse_output=False)
            self.labels = encoder.fit_transform(np.array(labels).reshape(-1, 1))
        else:
            self.labels = labels
    
    def __len__(self):
        return len(self.input_indices)
    
    def __getitem__(self, idx):
        indices = torch.tensor(self.input_indices[idx], dtype=torch.long)
        if self.one_hot:
            label = torch.tensor(self.labels[idx], dtype=torch.float)
        else:
            label = torch.tensor(self.labels[idx], dtype=torch.long)
        return indices, label

# 定義訓練函數
def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    for inputs, labels in tqdm(dataloader, desc="訓練中"):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)

# 定義驗證函數
def eval_epoch(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            total_loss += loss.item()
            if labels.dim() == 2:  # One-Hot
                preds = torch.argmax(outputs, dim=1)
                true = torch.argmax(labels, dim=1)
            else:  # Label Encoding
                preds = torch.argmax(outputs, dim=1)
                true = labels
            correct += (preds == true).sum().item()
            total += labels.size(0)
    accuracy = correct / total
    return total_loss / len(dataloader), accuracy

# 早停機制
class EarlyStopping:
    def __init__(self, patience=3, delta=0):
        self.patience = patience
        self.delta = delta
        self.best_score = None
        self.early_stop = False
        self.counter = 0
        self.best_loss = float('inf')
        self.best_model_state = None

    def __call__(self, val_loss, model):
        score = -val_loss
        if self.best_score is None:
            self.best_score = score
            self.best_loss = val_loss
            self.best_model_state = model.state_dict()
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.best_loss = val_loss
            self.best_model_state = model.state_dict()
            self.counter = 0

# 網格搜索參數
param_grid = {
    'expand_abbreviations': [True, False],
    'keep_usernames': [True, False],
    'keep_negation': [True, False],
    'use_attention': [True, False],
    'use_hidden_states': [True, False],
    'one_hot': [True, False],
    'max_length': [30, 50]
}

# 儲存結果
results = []
best_acc = 0.0
best_params = None
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'使用設備: {device}')

# 載入GloVe詞向量
GLOVE_FILE = r'C:\Users\skych\glove.twitter.27B\glove.twitter.27B.100d.txt'
print(f'載入GloVe詞向量從 {GLOVE_FILE}...')
if not os.path.exists(GLOVE_FILE):
    raise FileNotFoundError(f"GloVe file not found at {GLOVE_FILE}")
glove_vectors = load_glove_vectors(GLOVE_FILE)
EMBEDDING_DIM = 100
print(f'GloVe詞向量載入完成，詞彙量：{len(glove_vectors)}')

# 檢查並編碼Sentiment欄位
print("檢查train_df和test_df的欄位...")
if 'train_df' not in globals() or 'test_df' not in globals():
    raise NameError("train_df or test_df not defined. Please run Step 2 to load the data.")
required_columns = ['OriginalTweet', 'Sentiment']
for df_name, df in [('train_df', train_df), ('test_df', test_df)]:
    for col in required_columns:
        if col not in df.columns:
            raise KeyError(f"Column '{col}' not found in {df_name}. Available columns: {df.columns}")
sentiment_map = {
    'Extremely Negative': 0,
    'Negative': 1,
    'Neutral': 2,
    'Positive': 3,
    'Extremely Positive': 4
}
if train_df['Sentiment'].dtype == 'object':
    print("將Sentiment欄位編碼為數字...")
    if not all(val in sentiment_map for val in train_df['Sentiment'].unique()):
        raise ValueError(f"Unexpected Sentiment values in train_df: {train_df['Sentiment'].unique()}")
    if not all(val in sentiment_map for val in test_df['Sentiment'].unique()):
        raise ValueError(f"Unexpected Sentiment values in test_df: {test_df['Sentiment'].unique()}")
    train_df['Sentiment'] = train_df['Sentiment'].map(sentiment_map)
    test_df['Sentiment'] = test_df['Sentiment'].map(sentiment_map)
print("train_df Sentiment values:", train_df['Sentiment'].value_counts())
print("test_df Sentiment values:", test_df['Sentiment'].value_counts())

# 遍歷所有組合
combinations = list(product(*param_grid.values()))
for combo in tqdm(combinations, desc='Grid Search'):
    params = dict(zip(param_grid.keys(), combo))
    expand_abbr, keep_users, keep_neg, use_attn, use_hidden, one_hot, max_len = combo
    
    # 步驟3：資料清理
    train_df['clean_tokens'] = train_df['OriginalTweet'].apply(
        lambda x: clean_text(x, expand_abbreviations=expand_abbr, keep_usernames=keep_users, keep_negation=keep_neg))
    test_df['clean_tokens'] = test_df['OriginalTweet'].apply(
        lambda x: clean_text(x, expand_abbreviations=expand_abbr, keep_usernames=keep_users, keep_negation=keep_neg))
    
    # 步驟4：詞向量化和嵌入矩陣
    all_tokens = [token for tokens in train_df['clean_tokens'] for token in tokens]
    vocab = {word: idx + 1 for idx, word in enumerate(set(all_tokens))}
    vocab['<PAD>'] = 0
    VOCAB_SIZE = len(vocab)
    MAX_LENGTH = max_len
    train_df['input_indices'] = train_df['clean_tokens'].apply(lambda x: tokens_to_indices(x, vocab, max_length=MAX_LENGTH))
    test_df['input_indices'] = test_df['clean_tokens'].apply(lambda x: tokens_to_indices(x, vocab, max_length=MAX_LENGTH))
    embedding_matrix = get_embedding_matrix(vocab, glove_vectors, EMBEDDING_DIM).to(device)
    
    # 步驟5：準備 Dataset 和 DataLoader
    train_dataset = CommentDataset(train_df['input_indices'].tolist(), train_df['Sentiment'].tolist(), one_hot=one_hot)
    test_dataset = CommentDataset(test_df['input_indices'].tolist(), test_df['Sentiment'].tolist(), one_hot=one_hot)
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    # 計算類別權重（僅對 Label Encoding）
    if not one_hot:
        class_weights = compute_class_weight('balanced', classes=np.unique(train_df['Sentiment']), y=train_df['Sentiment'])
        class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)
        criterion = nn.CrossEntropyLoss(weight=class_weights)
    else:
        criterion = nn.BCEWithLogitsLoss()
    
    # 初始化模型
    model = BiLSTMClassifier(
        vocab_size=VOCAB_SIZE,
        embedding_dim=EMBEDDING_DIM,
        hidden_dim=256,
        output_dim=5,
        num_layers=4,
        dropout=0.2,
        use_hidden_states=use_hidden,
        use_attention=use_attn,
        embedding_matrix=embedding_matrix
    ).to(device)
    
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)
    early_stopping = EarlyStopping(patience=3)
    
    # 訓練模型
    train_losses, val_losses, val_accuracies = [], [], []
    for epoch in range(5):
        train_loss = train_epoch(model, train_loader, criterion, optimizer, device)
        val_loss, val_acc = eval_epoch(model, test_loader, criterion, device)
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        val_accuracies.append(val_acc)
        scheduler.step(val_loss)
        early_stopping(val_loss, model)
        if early_stopping.early_stop:
            break
    
    # 儲存結果
    results.append({
        'expand_abbreviations': expand_abbr,
        'keep_usernames': keep_users,
        'keep_negation': keep_neg,
        'use_attention': use_attn,
        'use_hidden_states': use_hidden,
        'one_hot': one_hot,
        'max_length': max_len,
        'val_accuracy': val_acc
    })
    
    # 更新最佳組合
    if val_acc > best_acc:
        best_acc = val_acc
        best_params = params
        torch.save(early_stopping.best_model_state, '0511_best_grid_model.pth')

# 輸出結果
results_df = pd.DataFrame(results)
print("\n所有組合的驗證集準確率：")
print(results_df)
print("\n最佳組合：")
print(f"參數：{best_params}")
print(f"驗證集準確率：{best_acc:.4f}")
print("最佳模型已儲存至 0511_best_grid_model.pth")

# 儲存結果到 CSV
results_df.to_csv('0511_grid_search_results.csv', index=False)
```

## 步驟六：建立BiLSTM 模型

### 泰源版本Training

```python
class BiLSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim, num_layers=4, dropout=0.2, use_hidden_states=False, use_attention=False):
        super(BiLSTMClassifier, self).__init__()
        self.hidden_dim = hidden_dim
        self.use_attention = use_attention
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.embedding.weight = nn.Parameter(embedding_matrix)
        self.embedding.weight.requires_grad = True
        self.dropout = nn.Dropout(dropout)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers=num_layers, 
                           bidirectional=True, batch_first=True)
        if use_attention:
            self.attention = nn.Linear(hidden_dim * 2, 1)
        # 修正：明確指定 batch_norm 的 num_features
        self.batch_norm = nn.BatchNorm1d(hidden_dim * 2)
        self.fc = nn.Linear(hidden_dim * 2, output_dim)
        self.use_hidden_states = use_hidden_states

    def forward(self, text):
        # 調試：打印輸入形狀
        # print(f"Input text shape: {text.shape}")
        embedded = self.dropout(self.embedding(text))  # [batch_size, max_length, embedding_dim]
        # print(f"Embedded shape: {embedded.shape}")
        lstm_out, (hidden, cell) = self.lstm(embedded)  # lstm_out: [batch_size, max_length, hidden_dim*2]
        # print(f"LSTM output shape: {lstm_out.shape}, Hidden shape: {hidden.shape}")
        
        if self.use_attention:
            attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
            context = torch.sum(attn_weights * lstm_out, dim=1)
            hidden = context
        elif self.use_hidden_states:
            hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)
        else:
            hidden = torch.mean(lstm_out, dim=1)  # 使用均值池化作為默認
        
        # 調試：打印 batch_norm 前的形狀
        # print(f"Hidden shape before batch_norm: {hidden.shape}")
        hidden = self.batch_norm(hidden)
        hidden = self.dropout(hidden)
        return self.fc(hidden)

# 設定模型參數
embedding_matrix = embedding_matrix.to(device)
OUTPUT_DIM = 5
HIDDEN_DIM = 256
NUM_LAYERS = 4
DROPOUT = 0.2
LEARNING_RATE = 0.001

# 創建全新模型，確保無舊狀態
model = BiLSTMClassifier(
    vocab_size=VOCAB_SIZE,
    embedding_dim=EMBEDDING_DIM,
    hidden_dim=HIDDEN_DIM,
    output_dim=OUTPUT_DIM,
    dropout=DROPOUT,
    num_layers=NUM_LAYERS,
    use_hidden_states=False,
    use_attention=True
).to(device)
print(model)

# 驗證 batch_norm 參數
print(f"BatchNorm num_features: {model.batch_norm.num_features}")  # 應為 512

# 定義損失函數和優化器
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2, verbose=True)

# 定義訓練函數
def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    for inputs, labels in tqdm(dataloader, desc="訓練中"):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)

# 定義驗證函數
def eval_epoch(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            total_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
    accuracy = correct / total
    return total_loss / len(dataloader), accuracy

# 早停機制
class EarlyStopping:
    def __init__(self, patience=3, delta=0):
        self.patience = patience
        self.delta = delta
        self.best_score = None
        self.early_stop = False
        self.counter = 0
        self.best_loss = float('inf')

    def __call__(self, val_loss):
        score = -val_loss
        if self.best_score is None:
            self.best_score = score
            self.best_loss = val_loss
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.best_loss = val_loss
            self.counter = 0

# 訓練主流程
EPOCHS = 20
train_losses = []
val_losses = []
val_accuracies = []
early_stopping = EarlyStopping(patience=3)

for epoch in range(EPOCHS):
    print(f"\n==== 第{epoch+1}輪訓練 ====")
    train_loss = train_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = eval_epoch(model, test_loader, criterion, device)
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    val_accuracies.append(val_acc)
    print(f"訓練損失={train_loss:.4f}，驗證損失={val_loss:.4f}，驗證準確率={val_acc:.4f}")
    
    scheduler.step(val_loss)
    early_stopping(val_loss)
    if early_stopping.early_stop:
        print(f"早停觸發，停止於第{epoch+1}輪")
        break
```
